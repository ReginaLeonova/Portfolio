```python
import seaborn as sns
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
life_expectancy = pd.read_csv("all_data.csv")
print(life_expectancy.head())
```

```python
#visualize trends over time (life expectancy)
sns.lineplot(data=life_expectancy, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life Expectancy Over Years')
plt.show()
#We can see that Zimbabwe has the lowest life expectancy rate,however after 2008 among all countries it showed he greatest increase. 
#Germany shows the highest and the most stable rate for life expectancy.
```

```python
#compare average life expectancyacross countries
countries = life_expectancy['Country']
years_expect = life_expectancy['Life expectancy at birth (years)']
plt.figure(figsize=(10, 6))  
plt.bar(countries, years_expect, color='skyblue')
plt.xlabel('Countries')
plt.ylabel('Average Life Expectancy (years)')
plt.title('Comparison of Average Life Expectancy Across Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
#visualize trends over time (GDP)
sns.lineplot(data=life_expectancy, x='Year', y='GDP', hue='Country')
plt.title('GDP Over Years')
plt.show()
#USA shows the highest GDP over years among other countries. China showed the highest increase among all countries
```

```python
#visualize the relationship between GDP and life expectancy
sns.lmplot(data=life_expectancy, x='GDP', y='Life expectancy at birth (years)', hue='Country',aspect=1.5)
plt.title('GDP vs Life Expectancy')
plt.show()
#In Germany,China and USA we can see slight dependency between higher GDP and higher expectancy of life.
#However, we can also see that high GDP doesn't guarantee high expectancy of life.
#In countriies like Chile, Mexico and Germany GDP is small, but life expectancy is quite high. 
```

```python
sns.histplot(data=life_expectancy, x='GDP')
plt.show()
#The most common GDP rate is between 0.00 and 0.25
```

```python
sns.histplot(data=life_expectancy, x='Life expectancy at birth (years)')
plt.show()
#The most common rate of life expectancy is around 78 years. 
```

```python
sns.lmplot(x='GDP', y='Life expectancy at birth (years)', hue='Country',
           data=life_expectancy[(life_expectancy['GDP'] > 1.50) & (life_expectancy['Life expectancy at birth (years)'] > 70)], 
           aspect=2.0)
plt.title('GDP vs Life Expectancy')
plt.show()
#In Germany,China and USA we can see slight dependency between higher GDP and higher expectancy of life.
```
